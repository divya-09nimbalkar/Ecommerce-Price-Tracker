# Ecommerce Price Tracker - Exploration Notebook

This notebook demonstrates how to:
- Load raw scraped product price data
- Clean and process it
- Analyze price trends
- Visualize results



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Load Raw Data
We start by loading the sample dataset from `data/raw/sample.csv`.


```python
df = pd.read_csv(r"D:\Projects all\All GITHUB PROJECTS TO UPLOAD\Z FOLDERS OF ALL THE MAIN TOPICS\Data_Engineering\Ecommerce_Price_Tracker\data\raw\sample.csv")
df.columns = df.columns.str.strip().str.lower()
df["date"] = pd.to_datetime(df["date"])
df.head()
```

## 2. Summary Statistics
Basic overview of the dataset.


```python
print("Rows:", len(df))
print("Columns:", df.columns.tolist())
print("Average Price:", df["price"].mean())
print("Average Volume:", df["volume"].mean())
```

## 3. Price Trend Analysis
Visualize how product prices change over time.


```python
plt.figure(figsize=(10,5))
plt.plot(df["date"], df["price"], marker="o")
plt.title("Price Trend")
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

## 4. Volume Trend Analysis
Visualize how sales volume changes over time.


```python
plt.figure(figsize=(10,5))
plt.bar(df["date"], df["volume"])
plt.title("Volume Trend")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.xticks(rotation=45)
plt.show()
```

## 5. Moving Average
Smooth out price fluctuations with a rolling average.


```python
df["price_ma"] = df["price"].rolling(window=3).mean()
plt.figure(figsize=(10,5))
plt.plot(df["date"], df["price"], marker="o", label="Price")
plt.plot(df["date"], df["price_ma"], marker="o", linestyle="--", label="3-day MA")
plt.title("Price with Moving Average")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

## 6. Volatility Analysis
Measure daily percentage changes in price.


```python
df["price_change_pct"] = df["price"].pct_change() * 100
plt.figure(figsize=(10,5))
plt.bar(df["date"], df["price_change_pct"])
plt.title("Daily Price Change (%)")
plt.xlabel("Date")
plt.ylabel("Change (%)")
plt.xticks(rotation=45)
plt.show()
```

## 7. Price vs Volume Relationship
Scatter plot to see how price relates to sales volume.


```python
plt.figure(figsize=(6,6))
plt.scatter(df["price"], df["volume"], c="blue", alpha=0.7)
plt.title("Price vs Volume")
plt.xlabel("Price")
plt.ylabel("Volume")
plt.show()
```

## 8. Correlation Heatmap
Visualize correlations between numeric features.


```python
df_numeric = df.select_dtypes(include=["number"])
plt.figure(figsize=(8,6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

## 9. Top 5 Products by Average Price
Rank products by their average price across all days.


```python
top5 = df.groupby("name")["price"].mean().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
top5.plot(kind="bar", color="teal")
plt.title("Top 5 Products by Average Price")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```
